# Some examples of how to import the theory data, all IAM

```python
import xrayscatteringtools as xrst
import matplotlib.pyplot as plt
import numpy as np
```

```python
print(open('Theory/info.txt').read())
```

## Cheng and Yong Geometries

```python
# Define q space to calculate scattering patters.
q = np.linspace(0,8,200)

total_ground = xrst.theory.iam_total_pattern('Theory/DMP_ground_Cheng-et-al.xyz',q)
total_3sL = xrst.theory.iam_total_pattern('Theory/3sD_Yong-et-al.xyz',q)
total_3sD = xrst.theory.iam_total_pattern('Theory/3sL_Yong-et-al.xyz',q)

elastic_ground = xrst.theory.iam_elastic_pattern('Theory/DMP_ground_Cheng-et-al.xyz',q)
elastic_3sL = xrst.theory.iam_elastic_pattern('Theory/3sD_Yong-et-al.xyz',q)
elastic_3sD = xrst.theory.iam_elastic_pattern('Theory/3sL_Yong-et-al.xyz',q)

inelastic_ground = xrst.theory.iam_inelastic_pattern('Theory/DMP_ground_Cheng-et-al.xyz',q)
inelastic_3sL = xrst.theory.iam_inelastic_pattern('Theory/3sD_Yong-et-al.xyz',q)
inelastic_3sD = xrst.theory.iam_inelastic_pattern('Theory/3sL_Yong-et-al.xyz',q)


plt.figure(figsize=[15,5])
plt.subplot(1,3,1)
plt.plot(q,total_ground,label='Ground')
plt.plot(q,total_3sL,label='3sD')
plt.plot(q,total_3sD,label='3sL')
plt.title('Total')
plt.xlabel('q')
plt.ylabel('I(q)')
plt.legend()

plt.subplot(1,3,2)
plt.plot(q,elastic_ground,label='Ground')
plt.plot(q,elastic_3sL,label='3sD')
plt.plot(q,elastic_3sD,label='3sL')
plt.title('Elastic')
plt.xlabel('q')
plt.ylabel('I(q)')
plt.legend()

plt.subplot(1,3,3)
plt.plot(q,inelastic_ground,label='Ground')
plt.plot(q,inelastic_3sD,label='3sD')
plt.plot(q,inelastic_3sL,label='3sL')
plt.title('Inelastic')
plt.xlabel('q')
plt.ylabel('I(q)')
plt.legend()

plt.suptitle('Cheng/Yong et. al. Geometries.')
plt.show()
```

## Percent Differences

```python
plt.figure(figsize=[15,5])
plt.subplot(1,3,1)
plt.plot(q,100*(total_ground-total_3sL)/total_ground,label='3sL')
plt.plot(q,100*(total_ground-total_3sD)/total_ground,label='3sD')
plt.title('Total')
plt.xlabel('q')
plt.ylabel('I(q)')
plt.legend()

plt.subplot(1,3,2)
plt.plot(q,100*(elastic_ground-elastic_3sL)/elastic_ground,label='3sL')
plt.plot(q,100*(elastic_ground-elastic_3sD)/elastic_ground,label='3sD')
plt.title('Elastic')
plt.xlabel('q')
plt.ylabel('I(q)')
plt.legend()

plt.subplot(1,3,3)
plt.plot(q,100*(inelastic_ground-inelastic_3sL)/inelastic_ground,label='3sL')
plt.plot(q,100*(inelastic_ground-inelastic_3sD)/inelastic_ground,label='3sD')
plt.title('Inelastic')
plt.xlabel('q')
plt.ylabel('I(q)')
plt.legend()

plt.suptitle('Cheng/Yong et. al. Geometries.')
plt.show()
```

## Alec's Geometries

```python
total_ground_alec = xrst.theory.iam_total_pattern('Theory/DMP_ground_Alec.xyz',q)
total_L1 = xrst.theory.iam_total_pattern('Theory/DMP_L1.xyz',q)
total_L2 = xrst.theory.iam_total_pattern('Theory/DMP_L2.xyz',q)
total_D = xrst.theory.iam_total_pattern('Theory/DMP_D.xyz',q)

elastic_ground_alec = xrst.theory.iam_elastic_pattern('Theory/DMP_ground_Alec.xyz',q)
elastic_L1 = xrst.theory.iam_elastic_pattern('Theory/DMP_L1.xyz',q)
elastic_L2 = xrst.theory.iam_elastic_pattern('Theory/DMP_L2.xyz',q)
elastic_D = xrst.theory.iam_elastic_pattern('Theory/DMP_D.xyz',q)

inelastic_ground_alec = xrst.theory.iam_inelastic_pattern('Theory/DMP_ground_Alec.xyz',q)
inelastic_L1 = xrst.theory.iam_inelastic_pattern('Theory/DMP_L1.xyz',q)
inelastic_L2 = xrst.theory.iam_inelastic_pattern('Theory/DMP_L2.xyz',q)
inelastic_D = xrst.theory.iam_inelastic_pattern('Theory/DMP_D.xyz',q)


plt.figure(figsize=[15,5])
plt.subplot(1,3,1)
plt.plot(q,total_ground_alec,label='Ground')
plt.plot(q,total_L1,label='L1')
plt.plot(q,total_L2,label='L2')
plt.plot(q,total_D,label='D')
plt.title('Total')
plt.xlabel('q')
plt.ylabel('I(q)')
plt.legend()

plt.subplot(1,3,2)
plt.plot(q,elastic_ground_alec,label='Ground')
plt.plot(q,elastic_L1,label='L1')
plt.plot(q,elastic_L2,label='L2')
plt.plot(q,elastic_D,label='D')
plt.title('Elastic')
plt.xlabel('q')
plt.ylabel('I(q)')
plt.legend()

plt.subplot(1,3,3)
plt.plot(q,inelastic_ground_alec,label='Ground')
plt.plot(q,inelastic_L1,label='L1')
plt.plot(q,inelastic_L2,label='L2')
plt.plot(q,inelastic_D,label='D')
plt.title('Inelastic')
plt.xlabel('q')
plt.ylabel('I(q)')
plt.legend()

plt.suptitle('Alec Geometries.')
plt.show()
```

```python
plt.figure(figsize=[15,5])
plt.subplot(1,3,1)
plt.plot(q,100*(total_ground_alec-total_L1)/total_ground_alec,label='L1')
plt.plot(q,100*(total_ground_alec-total_L2)/total_ground_alec,label='L2')
plt.plot(q,100*(total_ground_alec-total_D)/total_ground_alec,label='D')
plt.title('Total')
plt.xlabel('q')
plt.ylabel('I(q)')
plt.legend()

plt.subplot(1,3,2)
plt.plot(q,100*(elastic_ground_alec-elastic_L1)/elastic_ground_alec,label='L1')
plt.plot(q,100*(elastic_ground_alec-elastic_L2)/elastic_ground_alec,label='L2')
plt.plot(q,100*(elastic_ground_alec-elastic_D)/elastic_ground_alec,label='D')
plt.title('Elastic')
plt.xlabel('q')
plt.ylabel('I(q)')
plt.legend()

plt.subplot(1,3,3)
plt.plot(q,100*(inelastic_ground_alec-inelastic_L1)/inelastic_ground_alec,label='L1')
plt.plot(q,100*(inelastic_ground_alec-inelastic_L2)/inelastic_ground_alec,label='L2')
plt.plot(q,100*(inelastic_ground_alec-inelastic_D)/inelastic_ground_alec,label='D')
plt.title('Inelastic')
plt.xlabel('q')
plt.ylabel('I(q)')
plt.legend()

plt.suptitle('Alec Geometries.')
plt.show()
```
